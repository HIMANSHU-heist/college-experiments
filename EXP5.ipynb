```python
def sort_cities(input_file, output_file):
    
        with open(input_file, 'r') as f:
            cities = f.readlines()

        cities = [city.strip() for city in cities if city.strip()]
        cities.sort()

        with open(output_file, 'w') as f:
            for city in cities:
                f.write(city + '\n')
        
        print(f"Success! Sorted cities saved to {output_file}")

    

# Usage
in_file = "city.txt"
out_file = "sortedcity.txt"
sort_cities(in_file, out_file)
```
