```python
a = 10
b = 20.5
c = "Python"
d = True
e = [1, 2, 3, "Apple"]
f = (10, 20, 30)
g = {"name": "Amit", "age": 15}
h = {1, 2, 3, 4, 5}

print("Integer:", a, type(a))
print("Float:", b, type(b))
print("String:", c, type(c))
print("Boolean:", d, type(d))
print("List:", e, type(e))
print("Tuple:", f, type(f))
print("Dictionary:", g, type(g))
print("Set:", h, type(h))
```
