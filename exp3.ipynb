A)to manage a task using list and tuples  including adding removing and sorting task
b)use of set and set opraton to manage student enrollment in multiple entrance exam like cet,jee,neet
c)to create update and manuipulate the dictornary of student record inlcuding grades and attendence

```python
tasks = []

while True:
    name = input("Enter task name (or 'done' to stop): ")
    if name.lower() == 'done':
        break
    priority = int(input(f"Enter priority for '{name}': "))
    tasks.append((name, priority))

tasks.sort(key=lambda x: x[1])

print("\nList after adding and sorting:")
print(tasks)

while True:
    remove_name = input("\nEnter task name to remove (or 'skip' to finish): ")
    if remove_name.lower() == 'skip':
        break
    tasks = [t for t in tasks if t[0] != remove_name]
    print("Updated list:")
    print(tasks)

print("\nFinal Task List:")
print(tasks)
```

```python
jee = set(input("Enter JEE students (comma separated): ").split(","))
neet = set(input("Enter NEET students (comma separated): ").split(","))

both = jee & neet
either = jee | neet
only_jee = jee - neet

print("Both Exams:", both)
print("All Unique Students:", either)
print("Only JEE:", only_jee)
```

```python
records = {}

id = input("Enter Student ID: ")
name = input("Enter Name: ")
grade = input("Enter Grade: ")
attendance = int(input("Enter Attendance %: "))

records[id] = {"name": name, "grade": grade, "attendance": attendance}

up_id = input("Enter ID to update grade: ")
new_grade = input("Enter new grade: ")
if up_id in records:
    records[up_id]["grade"] = new_grade

print(records)
```
