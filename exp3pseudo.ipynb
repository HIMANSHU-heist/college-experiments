```python
tasks = []

# Adding tasks
n = int(input("How many tasks? "))
for i in range(n):
    t = input(f"Enter task {i+1}: ")
    tasks.append(t)

print("Current Tasks:", tasks)

# Removing a task
rem = input("Enter task to remove: ")
if rem in tasks:
    tasks.remove(rem)

# Sorting tasks
tasks.sort()

# Updating a task
idx = int(input("Enter index to update (0, 1...): "))
tasks[idx] = input("Enter new task: ")

# Final Tuple
final_tasks = tuple(tasks)
print("Final Tasks (Tuple):", final_tasks)
```

```python
# JEE Students
jee_set = set()
n1 = int(input("Number of JEE students: "))
for i in range(n1):
    jee_set.add(input("Enter name: "))

# NEET Students
neet_set = set()
n2 = int(input("Number of NEET students: "))
for i in range(n2):
    neet_set.add(input("Enter name: "))

print("\nJEE List:", jee_set)
print("NEET List:", neet_set)

# Operations
print("Total unique students (Union):", jee_set | neet_set)
print("Students in both exams (Intersection):", jee_set & neet_set)
print("Only in JEE (Difference):", jee_set - neet_set)
```

```python
records = {}

n = int(input("How many records? "))
for i in range(n):
    name = input("Enter Student Name: ")
    grade = input("Enter Grade: ")
    att = input("Enter Attendance %: ")
    records[name] = {"Grade": grade, "Attendance": att}

print("\nInitial Database:", records)

# Updating
up_name = input("\nName to update grade: ")
if up_name in records:
    records[up_name]["Grade"] = input("New Grade: ")

# Deleting
del_name = input("Name to delete: ")
records.pop(del_name, None)

print("Final Database:", records)
```
