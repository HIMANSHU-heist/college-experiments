```python
def calculator(num1,num2,choice):
    if choice==1:
        return num1 + num2
    elif choice==2:
        return num1-num2
    elif choice==3:
        if num1 or num2 ==0:
            print(" plz enter valid number ")
        else:
            return num1/num2
    elif choice==4:
        return num1*num2
    else:
        print("plz enter valid operation")

print("1=addition,2=substraction,3=division,4=multiplication")
A=int(input("enter num1 : "))
B=int(input("enter num2 : "))
C=int(input("enter OPERATION(1,2 ETC,.) : "))
RESULT=calculator(A,B,C)
print(RESULT)
```

```python
def check_even_odd(n):
    if n % 2 == 0:
        print(f"{n} is Even")
    else:
        print(f"{n} is Odd")

def find_factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter a number: "))
check_even_odd(num)

if num < 0:
    print("Factorial not possible for negative numbers.")
else:
    print(f"Factorial of {num} is: {find_factorial(num)}")9
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter number to check prime: "))

if is_prime(num):
    print(f"{num} is a Prime Number")
else:
    print(f"{num} is not a Prime Number")
```
