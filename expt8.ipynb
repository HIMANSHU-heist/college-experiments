```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.__account_number = account_number
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance -= amount

    def check_balance(self):
        return self.__balance


account = BankAccount('12345678', 1000)
account.deposit(500)
account.withdraw(200)
print(account.check_balance())
```

```python
from datetime import datetime

class Vehicle:
    def __init__(self, vehicle_id, model, daily_rate):
        self.vehicle_id = vehicle_id
        self.model = model
        self.daily_rate = daily_rate
        self.is_available = True

    def __str__(self):
        status = "Available" if self.is_available else "Rented"
        return f"[{self.vehicle_id}] {self.model} - ${self.daily_rate}/day ({status})"

class RentalTransaction:
    def __init__(self, customer_name, vehicle, start_date, end_date):
        self.customer_name = customer_name
        self.vehicle = vehicle
        self.start_date = datetime.strptime(start_date, "%Y-%m-%d")
        self.end_date = datetime.strptime(end_date, "%Y-%m-%d")
        
    def calculate_duration(self):
        duration = (self.end_date - self.start_date).days
        return max(duration, 1)  # Minimum 1 day charge

    def calculate_total_price(self):
        return self.calculate_duration() * self.vehicle.daily_rate

    def confirm_booking(self):
        self.vehicle.is_available = False
        return f"Booking confirmed for {self.customer_name}: {self.vehicle.model} for {self.calculate_duration()} days."

class RentalAgency:
    def __init__(self, name):
        self.name = name
        self.inventory = []

    def add_vehicle(self, vehicle):
        self.inventory.append(vehicle)

    def get_available_vehicles(self):
        return [v for v in self.inventory if v.is_available]

    def book_vehicle(self, customer_name, vehicle_id, start_date, end_date):
        # Find vehicle by ID
        vehicle = next((v for v in self.inventory if v.vehicle_id == vehicle_id), None)
        
        if vehicle and vehicle.is_available:
            transaction = RentalTransaction(customer_name, vehicle, start_date, end_date)
            print(transaction.confirm_booking())
            print(f"Total Cost: ${transaction.calculate_total_price()}")
            return transaction
        else:
            print("Error: Vehicle not available or ID not found.")
            return None


agency = RentalAgency("Bendalehimanshu")
agency.add_vehicle(Vehicle("C001", "Tesla Model 3", 120))
agency.add_vehicle(Vehicle("C002", "Toyota supra", 50))


print(f"Welcome to {agency.name}! Available cars:")
for car in agency.get_available_vehicles():
    print(car)


print("Processing Booking")
agency.book_vehicle("himanshu bendale", "C001", "2026-04-01", "2026-04-05")


print("\nUpdated Inventory:")
for car in agency.inventory:
    print(car)
```

```python
class Participant:
    def __init__(self, name, email):
        self.name = name
        self.email = email

class Activity:
    def __init__(self, title, time):
        self.title = title
        self.time = time

class Event:
    def __init__(self, event_name, date):
        self.event_name = event_name
        self.date = date
        self.activities = []    # List to store Activity objects
        self.participants = []  # List to store Participant objects

    def add_activity(self, activity_obj):
        self.activities.append(activity_obj)
        print(f"Activity '{activity_obj.title}' added to {self.event_name}.")

    def register_participant(self, participant_obj):
        self.participants.append(participant_obj)
        print(f"Registered: {participant_obj.name} for {self.event_name}")

    def show_details(self):
        print(f"\n--- {self.event_name} Details ---")
        print(f"Date: {self.date}")
        print(f"Total Participants: {len(self.participants)}")
        print("Schedule:")
        for act in self.activities:
            print(f" - {act.time}: {act.title}")




my_fest = Event("Annual College TechFest", "2026-05-20")


coding_war = Activity("Python Coding War", "10:00 AM")
gaming_zone = Activity("Esports Tournament", "02:00 PM")

my_fest.add_activity(coding_war)
my_fest.add_activity(gaming_zone)


p1 = Participant("John Doe", "john@example.com")
p2 = Participant("Sarah Lee", "sarah@example.com")

my_fest.register_participant(p1)
my_fest.register_participant(p2)


my_fest.show_details()
```
