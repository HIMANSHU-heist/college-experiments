```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.__account_number = account_number
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance -= amount

    def check_balance(self):
        return self.__balance


account = BankAccount('50000000', 45000)
account.deposit(100)
account.withdraw(400)
print(account.check_balance())
```

```python
class Vehicle:
    def __init__(self, name, price_per_day):
        self.name = name
        self.price_per_day = price_per_day
        self.is_available = True  # Start as available

class RentalAgency:
    def __init__(self):
        self.cars = [] # A list to hold our Vehicle objects

    def add_car(self, car):
        self.cars.append(car)

    def show_available_cars(self):
        print("\n--- Available Cars ---")
        for car in self.cars:
            if car.is_available:
                print(f"{car.name} - ${car.price_per_day} per day")

    def rent_car(self, car_name, days):
        for car in self.cars:
            if car.name == car_name and car.is_available:
                # Calculate the cost
                total_cost = car.price_per_day * days
                # Mark as rented
                car.is_available = False
                print(f"\nSuccess! You rented the {car.name} for {days} days.")
                print(f"Total cost: ${total_cost}")
                return
        
        print("\nSorry, that car is not available.")




my_agency = RentalAgency()

my_agency.add_car(Vehicle("fortuner", 50))
my_agency.add_car(Vehicle("scorpo", 100))


my_agency.show_available_cars()


my_agency.rent_car("scorpo", 3)


my_agency.show_available_cars()
```

```python
class Participant:
    def __init__(self, name, email):
        self.name = name
        self.email = email

class Activity:
    def __init__(self, title, time):
        self.title = title
        self.time = time

class Event:
    def __init__(self, event_name, date):
        self.event_name = event_name
        self.date = date
        self.activities = []    # List to store Activity objects
        self.participants = []  # List to store Participant objects

    def add_activity(self, activity_obj):
        self.activities.append(activity_obj)
        print(f"Activity '{activity_obj.title}' added to {self.event_name}.")

    def register_participant(self, participant_obj):
        self.participants.append(participant_obj)
        print(f"Registered: {participant_obj.name} for {self.event_name}")

    def show_details(self):
        print(f"\n--- {self.event_name} Details ---")
        print(f"Date: {self.date}")
        print(f"Total Participants: {len(self.participants)}")
        print("Schedule:")
        for act in self.activities:
            print(f" - {act.time}: {act.title}")

# --- Using the System ---

# 1. Create an Event
my_fest = Event("Annual College TechFest", "2026-05-20")

# 2. Add Activities (Scheduling)
coding_war = Activity("Python Coding War", "10:00 AM")
gaming_zone = Activity("Esports Tournament", "02:00 PM")

my_fest.add_activity(coding_war)
my_fest.add_activity(gaming_zone)

# 3. Register Participants
p1 = Participant("John Doe", "john@example.com")
p2 = Participant("Sarah Lee", "sarah@example.com")

my_fest.register_participant(p1)
my_fest.register_participant(p2)

# 4. Show the Full Plan
my_fest.show_details()
```
